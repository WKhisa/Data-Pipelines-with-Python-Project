```python
# Import libraries and fetch data
import pandas as pd
```

```python
#Load data
dataset1_df = pd.read_csv('dataset1.csv')
dataset2_df = pd.read_csv('dataset2.csv')
dataset3_df = pd.read_csv('dataset3.csv')
```

**Merge the datasets**

```python
combined_datasets_df = pd.merge(left=dataset1_df, right=dataset2_df, how='left', left_on=['customer_id','payment_status','payment_method'], right_on= ['customer_id','payment_status','payment_method'])
combined_datasets_df = pd.merge(left=combined_datasets_df, right=dataset3_df, how='left', left_on=['customer_id'], right_on= ['customer_id'])

combined_datasets_df
```

```python
combined_datasets_df.dropna(inplace= True)
```

**Save data as csv**

```python
combined_datasets_df.to_csv('final_billing_records.csv', header=True, index=False)
```
